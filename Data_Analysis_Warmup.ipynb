## Apply Method

```python
# using Normal Function

def marks(x): #function created
  return x/2

df1['Marks'].apply(marks) #function applied to marks column of df1
```

```python
# new column is added
df1['Half_marks']= df1['Marks'].apply(marks)
df1
```

----
----

```python
# Using Lambda Function

df1['Marks'].apply(lambda x: x/2)
```

```python
df1['lambda_marks']= df1['Marks'].apply(lambda x: x/2)
df1
```

```python
# use apply and length to find length of the string
df1['Name'].apply(len)
```

----
----

```python
# Map Fuction
#encoding Male-1 and Female-0
df1['Male_Female']= df1['Gender'].map({'Male':1, 'Female':0})
df1
```

----

```python
# Drop Column
df1.drop('Male_Female', axis=1)
```

```python
# Drop Multiple Columns

df1.drop(['Male_Female', 'lambda_marks', 'Half_marks'], axis=1, inplace=True)
```
